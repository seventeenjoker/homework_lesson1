# Остановки
Считать из csv-файла (с http://data.mos.ru/datasets/752) количество остановок, вывести улицу, на которой больше всего остановок.

```python
!ls
```

```python
import csv
```

```python
streets = {}
```

```python
%time
with open('bus_stops.csv', 'r', encoding='cp1251') as f:
    reader = csv.DictReader(f, delimiter=';')
    for row in reader:
        if row['Street'] in streets:
            streets[row['Street']] += 1
        else:
            streets[row['Street']] = 1
```

```python
streets
```

```python
%time
sorted(streets.items(), key=lambda item: item[1], reverse=True)
```

```python
show_first = sorted(streets.items(), key=lambda item: item[1], reverse=True)
```

```python
print(show_first[0])
```

```python
print(show_first[0][0], show_first[0][1])
```

```python
street = show_first[0][0]
```

```python
amount_stops = show_first[0][1]
```

```python
print(f'Улица, на которой больше всего остановок: {street}({amount_stops}).')
```
